```python
from liveml.object_detection.kitti.detector import Detector
import imageio
import os
from liveml.object_detection import kitti
from matplotlib import pyplot as plt
import numpy as np
```

```python
sample_location = kitti.__file__[:-11] + 'test_data/images/'
images = [imageio.imread(sample_location + file) for file in os.listdir(sample_location)]
```

```python
detector = Detector()
detector.init_session()
```

```python
%matplotlib inline

for detected_image in detector.visualize_detection_on_images(*images):
    plt.figure()
    plt.imshow(detected_image)
```
